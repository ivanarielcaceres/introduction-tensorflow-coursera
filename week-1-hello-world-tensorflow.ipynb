```
from tensorflow.keras import layers
from tensorflow import keras 
import tensorflow as tf
import numpy as np
```

Create the NN architecture and set the optimizer and loss function

```
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
```

Create data

```
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
print(type(xs), xs.shape)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
print(type(ys), ys.shape)
```

Fit the data to the model and set 500 iterations or epochs

```
model.fit(xs, ys, epochs=500)
```

Make some predictions

```
print(model.predict([10.0, 25.0, 50]))
```
